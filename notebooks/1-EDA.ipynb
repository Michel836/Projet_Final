```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration des graphiques
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Dossier d'entrée et de sortie
input_dir = "../data/raw/"
output_dir = "../data/processed/"
os.makedirs(output_dir, exist_ok=True)

# Liste des fichiers à traiter
files = {
    "train": "train (2).csv",
    "test": "test (1).csv",
    "evaluation": "evaluation.csv"
}

# Fonction de nettoyage des datasets
def clean_dataset(file_path, sep=";"):
    # Charger le dataset avec le bon séparateur
    data = pd.read_csv(file_path, sep=sep)
    
    # Supprimer les colonnes inutiles (colonnes "Unnamed")
    data_cleaned = data.drop(columns=[col for col in data.columns if "Unnamed" in col], errors="ignore")
    
    # Supprimer les lignes avec des valeurs manquantes dans les colonnes clés
    if all(col in data_cleaned.columns for col in ["id", "title", "text", "label"]):
        data_cleaned = data_cleaned.dropna(subset=["id", "title", "text", "label"])
    
    return data_cleaned

# Nettoyer et sauvegarder chaque fichier
for name, filename in files.items():
    print(f"Nettoyage du fichier {filename}...")
    file_path = os.path.join(input_dir, filename)
    
    # Nettoyer le dataset
    cleaned_data = clean_dataset(file_path)
    
    # Aperçu du dataset nettoyé
    print(f"\nAperçu des données nettoyées pour {name} :")
    print(cleaned_data.head())
    print(f"\nDimensions après nettoyage ({name}) : {cleaned_data.shape}")
    
    # Vérifier la distribution des classes
    if "label" in cleaned_data.columns:
        print(f"\nDistribution des classes dans {name} :")
        sns.countplot(x="label", data=cleaned_data)
        plt.title(f"Distribution des classes ({name})")
        plt.show()
    
    # Sauvegarder le dataset nettoyé
    output_path = os.path.join(output_dir, f"{name}_cleaned.csv")
    cleaned_data.to_csv(output_path, index=False)
    print(f"Dataset nettoyé sauvegardé sous : {output_path}")

print("\nEDA finalisé. Tous les fichiers nettoyés sont sauvegardés dans '../data/processed/'.")
```

```python
# Importation des bibliothèques
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuration des graphiques
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Chargement des fichiers nettoyés
train_data = pd.read_csv("../data/processed/train_cleaned.csv")
test_data = pd.read_csv("../data/processed/test_cleaned.csv")
evaluation_data = pd.read_csv("../data/processed/evaluation_cleaned.csv")

# Dimensions des datasets
print("Dimensions des datasets nettoyés :")
print(f"Train : {train_data.shape}")
print(f"Test : {test_data.shape}")
print(f"Evaluation : {evaluation_data.shape}")

# Vérification des classes
datasets = {"train": train_data, "test": test_data, "evaluation": evaluation_data}

for name, data in datasets.items():
    print(f"\nDistribution des classes dans {name} :")
    print(data["label"].value_counts(normalize=True))  # Proportions des classes
    sns.countplot(x="label", data=data)
    plt.title(f"Distribution des classes ({name})")
    plt.show()

# Analyse de la longueur des textes
for name, data in datasets.items():
    print(f"\nAnalyse des longueurs de texte dans {name} :")
    data["text_length"] = data["text"].apply(len)  # Longueur de chaque texte
    print(data["text_length"].describe())  # Statistiques descriptives
    sns.histplot(data["text_length"], bins=50, kde=True)
    plt.title(f"Distribution des longueurs de texte ({name})")
    plt.xlabel("Longueur du texte")
    plt.ylabel("Fréquence")
    plt.show()

# Vérification des valeurs uniques dans 'label'
for name, data in datasets.items():
    print(f"\nValeurs uniques dans 'label' ({name}) : {data['label'].unique()}")
```

```python
# Identifier les lignes avec la valeur incorrecte dans 'label'
print("Lignes problématiques dans train :")
print(train_data[train_data["label"] == " rejects capitalism as an nonviable economic system"])

print("\nLignes problématiques dans evaluation :")
print(evaluation_data[evaluation_data["label"] == " rejects capitalism as an nonviable economic system"])
```

```python
# Supprimer les lignes problématiques dans train et evaluation
train_data = train_data[train_data["label"] != " rejects capitalism as an nonviable economic system"]
evaluation_data = evaluation_data[evaluation_data["label"] != " rejects capitalism as an nonviable economic system"]
```

```python
# Convertir 'label' en entier pour uniformiser les données
train_data["label"] = train_data["label"].astype(int)
evaluation_data["label"] = evaluation_data["label"].astype(int)
```
